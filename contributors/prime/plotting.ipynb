```python
import pandas as pd
import geopandas as gpd
import json
from datetime import date, datetime
import matplotlib.pyplot as plt
```

```python
# read data
fname = "../beau-uriona/one_obj_to_rule_them_all.json"
df = gpd.read_file(fname)
df.head()
```

```python
snotel_variable = "depth_awdb"
snowex_variable = "depth_snowex"
```

```python
df["Depth Difference"] = df[snotel_variable] - df[snowex_variable]
```

```python
ax = df.plot.scatter(snotel_variable, snowex_variable, c='distance', legend=True, cmap='viridis')
# Set the various plots x/y labels and title.
ax.set_title(f'Depth Comparison')
ax.set_xlabel('AWDB Depth')
ax.set_ylabel('Snowex Depth')
```

```python
df.plot.scatter("distance", "Depth Difference", c=snotel_variable, cmap="viridis")
# ax.set_title(f'Depth Comparison')
ax.set_xlabel('Distance')
ax.set_ylabel('Depth')
```
