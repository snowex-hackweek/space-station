```python
import geopandas as gpd
import requests
url = "https://api.snowdata.info/stations/getMeta?siteId=ALL&network=ALL&state=ALL&element=ALL&meta=true&active=true&format=geojson&orient=records"
req = requests.get(url)

sntl_json = req.json()

gdf = gpd.GeoDataFrame.from_features(sntl_json)
gdf = gdf[gdf["beginDate"].str.startswith("2010")]
gdf.shape

```
